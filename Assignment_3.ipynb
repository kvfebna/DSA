## Who scores the most on average for math, reading and writing based on gender

```python
df.groupby('gender')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title('Math, Reading and Writing Score Averages based on Gender')
plt.ylabel('average')
plt.show()
```

OBSERVATION:
    
1. Male students scored more than female students on math average.
2. Average reading score of female students is greater than that of male students.
3. As the same as reading score average, Female students acquired high writing score average than male students.

## Who scores the most on average for math, reading and writing based on test preparation course

```python
df.groupby('test preparation course')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title('Math, Reading and Writing Score Averages based on Test Preparation Course')
plt.ylabel('average')
plt.show()
```

OBSERVATION: The averages of math, reading and writing scores are higher for the students who completed test preparation course compared to the students who does not completed it. So it is clear that, the test preparation course has a great impact on students academic performance.

## Scoring variation for math, reading and writing based on gender

### Maximum value, minimum value and variance 

```python
df.groupby('gender')[['math score','reading score','writing score']].agg(['max','min','var'])
```

OBSERVATION:
1. Range of female students math score = max-min = 100-0 = 100. 

   Range of male students math score = max-min = 100-27 = 73. 
   
   Also variance of math score is greater for female students compared to male students. Hence we can say that the math score variation is higher for female students than that of male students.
2. In the case of reading and writing score variations, male and female students are approximately same. But still female students possess a slighter high variation more than male students.

### Box plot

```python
fig,axes = plt.subplots(1, 3, sharex=True,
                            figsize=(13, 4))
sns.boxplot(x=df.gender,y=df['math score'],ax=axes[0])
sns.boxplot(x=df.gender,y=df['reading score'],ax=axes[1])
sns.boxplot(x=df.gender,y=df['writing score'],ax=axes[2])
plt.subplots_adjust(wspace=0.4)
plt.show()
```

OBSERVATION: Female students has more no.of outliers than that of male students in all the three scores.

### Skewness value 

```python
df.groupby('gender')[['math score','reading score','writing score']].skew()
```

OBSERVATION:

1. Math score skewness values of both female and male students are lie in between -0.5 and +0.5. So it indicates an approximately symmetric distribution.
2. In the case of reading score also, skewness values of both female and male students are lie in between -0.5 and +0.5. So this also indicates an approximately symmetric distribution.
3. The writing score skewness value of male students also indicates an approximately symmetric distribution. But for female students, writing score skewness value is -0.559205 which indicates a moderately skewed distribution.

### Plot distribution

```python
fig,axes = plt.subplots(1, 3, sharex=True, figsize=(13, 4))
plt.suptitle('Kernel Density Estimate (KDE) Plot of Math, Reading and Writing Scores based on Gender')
sns.kdeplot(df, x="math score", hue="gender", ax=axes[0])
sns.kdeplot(df, x="reading score", hue="gender", ax=axes[1])
sns.kdeplot(df, x="writing score", hue="gender", ax=axes[2])
plt.subplots_adjust(wspace=0.4)
plt.show()
```

## Find out top 25% of students based on their math score

Find the students who scored 77 or more in maths and sort them in descending order based on the math score
